```python
import classy as cl
from classy import Class
import numpy as np
import matplotlib.pyplot as plt
import math
import pickle
pi = np.pi
```

```python
VolGpc=200.
kmax=0.7
kMINGpc=2*pi/float(VolGpc)**(1/3.) #h^{-1} Gpc kmin
kMIN = kMINGpc*10**(-3)
kw=np.arange(kMIN,kmax,kMIN)

def Deriv_Pk_simple(para,steps,k):
    DPk_sim={}
    DeltaPk_P=np.zeros(len(k))
    DeltaPk_M=np.zeros(len(k))
    for key in steps.keys():
        cosmo = Class()
        paramstemp=para.copy()
        paramstemp[key]+=steps[key]
        cosmo.set(paramstemp)
        cosmo.compute()
        h=cosmo.h()
        for i in range(len(k)):
            DeltaPk_P[i]=cosmo.pk(h*k[i],0)*h**3
        paramstemp[key]-=2*steps[key]
        cosmo.set(paramstemp)
        cosmo.compute()
        h=cosmo.h()
        for i in range(len(k)):
            DeltaPk_M[i]=cosmo.pk(h*k[i],0)*h**3
        DPk_sim[key]=(DeltaPk_P-DeltaPk_M)/(2*steps[key])
        cosmo.struct_cleanup()
        cosmo.empty()
    return DPk_sim

def Fisher_Pk_simple(VG,Ng,kw,PkW,DPk,Dkeys,zavg,kmax):
    Fisher_sim=np.zeros((len(Dkeys),len(Dkeys)))
    aux={'vol':VG,'N_g':Ng,'z':zavg,'k_max':kmax}
    kMinGpc=2*pi/VG**(1/3.) #Gpc kmi
    V=float(VG*(10**3)**3)
    ndens=float(Ng)/V
    kMin = kMinGpc*10**(-3)
#     kMax=kmax
#     if kmax>kw.max():
#         kMax=kw.max()
#     endpoint=np.argwhere(kw<kMax).max()
    print(kMin)
    print(ndens)
    for i in range(len(Dkeys)):
        for j in range(len(Dkeys)):
            for p in range(len(kw)):
                Fisher_sim[i,j]+=(V/(4*pi**2))*kw[p]**2*kMin*(PkW[p]+1./(ndens))**(-2.)*DPk[Dkeys[i]][p]*DPk[Dkeys[j]][p]
                #Fisher_sim[i,j]+=(V/(4*pi**2))*kw[p]**2*kMin*(Dt(zavg)*Dt(zavg)*PkW[p]+1./(ndens))**(-2.)*DPk[Dkeys[i]][p]*DPk[Dkeys[j]][p]

    return Fisher_sim,aux
```

```python
def quick_Pk_zbins(VolGpc, z_bins, k_MAX=0.7, add={'N_ur':3.046}):
    z_string=''
    for elem in z_bins[:-1]:
        z_string+=str(elem)+','
    z_string+=str(z_bins[-1])
    k_MIN_Gpc=2*np.pi/float(VolGpc)**(1/3.)
    k_MIN = k_MIN_Gpc*10**(-3)
    
    k_array = np.arange(k_MIN, k_MAX, k_MIN)
    #define a dictionary with the parameters of desired cosmology
    params = {
            'output': 'tCl pCl lCl mPk',
            'l_max_scalars': 5000,
            'lensing': 'yes',
            'A_s': 2.1e-9,
            'n_s': 0.9667,
            '100*theta_s': 1.04112,
            'omega_b':0.02230,
            'omega_cdm': 0.1188,
            'tau_reio': 0.066,
            'N_ur': 3.046
    }
#     params = {
#         'output': 'tCl pCl lCl mPk',
#         'l_max_scalars': 5000,
#         'lensing': 'yes',
#         'A_s': 2.3e-9,
#         'n_s': 0.9624,
#         '100*theta_s': 1.042143,
#         'omega_b': 0.022068,
#         'omega_cdm': 0.12029,
#         'tau_reio':0.06,
#         'N_ur': 3.046
#     }
    for key in add.keys():
        params[key] = add[key]
    params['z_pk'] = z_string
    Pk = {}
    cosmo = Class()
    cosmo.set(params)
    cosmo.compute()
    h = cosmo.h()
    #there's no xrange in Python3
    for z_k in z_bins:
        Pk[z_k] = np.zeros(len(k_array))
        for i in range(len(k_array)):
            Pk[z_k][i] = cosmo.pk(h*k_array[i], z_k)*h**3
    cosmo.struct_cleanup()
    cosmo.empty()
    return Pk,k_array,params
```

```python
step_sizes = {
    'A_s': 1e-10,
    'N_ur': 0.06,
    'tau_reio': 0.02,
    'n_s': 0.01,
    '100*theta_s': 0.002,
    'omega_b': 0.0008,
    'omega_cdm': 0.002
}
parameters = {
        'output': 'tCl pCl lCl mPk',
        'l_max_scalars': 5000,
        'lensing': 'yes',
        'A_s': 2.1e-9,
        'n_s': 0.9667,
        '100*theta_s': 1.04112,
        'omega_b':0.02230,
        'omega_cdm': 0.1188,
        'tau_reio': 0.066,
        'N_ur': 3.046
}
```

```python
DPk_sim = Deriv_Pk_simple(parameters, step_sizes, kw)
print(DPk_sim)
```
